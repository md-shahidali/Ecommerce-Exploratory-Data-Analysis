## 📊 Univariate Analysis

```python
#Distribution of sales 

plt.figure(figsize=(10,6))
plt.hist(df["Sales"],color= "blue",bins=50, edgecolor= "black")
plt.title("Distribution of sales")
plt.xlabel("sales")
plt.ylabel("frequency")
plt.xlim(0,5000)
plt.tight_layout()
plt.savefig('../images/distributionOfSALES.png')
plt.show()
```

```python
#boxplot of profits for outlier detection

plt.boxplot(df['Profit'])
plt.title('Boxplot of Profit')
plt.ylabel('Profit')
plt.tight_layout()
plt.savefig('../images/profit_boxplot.png')
plt.show()
```

```python
#summary statistics of quantity 

df["Quantity"].describe()
```

```python
#percentage of (customers/orders) accross region for marketing 

regions = df["Region"].value_counts()
plt.pie(regions, labels=regions.index, autopct="%1.1f%%", startangle=90)
plt.title("Customers accross Regions")
plt.legend()
plt.tight_layout()
plt.savefig('../images/cus_region.png')
plt.show()
```

## 📊 Bivariate Analysis

```python
# Total Sales by category

category_sales = df.groupby("Category")["Sales"].sum().plot(kind="bar", color="orange", edgecolor="black")
plt.title("Total Sales by category")
plt.xlabel("category")
plt.ylabel("Sales")
plt.tight_layout()
plt.savefig('../images/salesBYcategory.png')
plt.show()
```

```python
#Realtionship between sales and profit

plt.figure(figsize=(8, 5))
plt.scatter(df["Sales"],df["Profit"], color="green",alpha=0.6)

plt.title("Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('../images/salesVSprofit.png')
plt.show()
```

```python
#relationship between discount and profit//

plt.figure(figsize=(10,6))
plt.scatter(df["Discount"], df["Profit"], color="orange", alpha=0.6, edgecolor="black")
plt.title("Relationship between Discount and Profit")
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('../images/dis_vs_profit.png')
plt.show()
```

```python
# Profit over category
category_profit = df.groupby("Category")["Profit"].sum().plot(kind="bar", color="purple", edgecolor="black")
plt.title("Profit over category")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.tight_layout()
plt.savefig('../images/profit_category.png')
plt.show()
```

```python
# profit and loss over regions

colors = ['green' if val >= 0 else 'red' for val in df['Profit']]

plt.bar(df['Region'], df['Profit'], color=colors)

plt.axhline(0, color='black', linewidth=1)

plt.title("Profit and loss over region")
plt.xlabel("Region")
plt.ylabel("Profit and loss")
plt.tight_layout()
plt.savefig('../images/profit_loss.png')
plt.show()
```

## 📊 Multivariate Analysis

```python
#average profit per region per category
avg_profit = df.groupby(["Region", "Category"])["Profit"].mean().unstack()

avg_profit.plot(kind="bar", figsize=(10,6))
plt.title("Average Profit by Category and Region")
plt.xlabel("Region")
plt.ylabel("Average Profit")
plt.tight_layout()
plt.savefig('../images/avg_profit.png')
plt.show()
```

## 📅 Time Series Analysis and Feature engineering

```python
# conversion of object to datetime 
df["Order Date"]=pd.to_datetime(df["Order Date"])
df.info() # for review
```

```python
#Feature engineering //

df["order day"] = df["Order Date"].dt.day
df["order Month"] = df["Order Date"].dt.month
df["order Year"] = df["Order Date"].dt.year
```

```python
# Monthly Sales

monthly_sales = df.groupby("order Month")[["Sales"]].sum().sort_index()
monthly_sales.plot(kind="bar", label="monthly sales")
plt.legend()
plt.title("Sales per month")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(axis="y",linestyle='--', alpha=0.6)
plt.xticks(ticks=range(0,12), labels=[
    'Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
plt.tight_layout()
plt.savefig('../images/monthly_sales.png')
plt.show()
```

```python
# sales vs profit overtime

monthly_data = df.groupby(['order Year', 'order Month'])[['Sales', 'Profit']].sum().reset_index()
monthly_data['Year-Month'] = pd.to_datetime(monthly_data['order Year'].astype(str) + '-' + monthly_data['order Month'].astype(str))
monthly_data = monthly_data.sort_values('Year-Month')

plt.figure(figsize=(14,6))
plt.plot(monthly_data['Year-Month'], monthly_data['Sales'], label='Sales', marker='o')
plt.plot(monthly_data['Year-Month'], monthly_data['Profit'], label='Profit', marker='s')
plt.title('Sales vs Profit Over Time')
plt.xlabel('Date')
plt.ylabel('Amount')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('../images/sales_vs_profit_overtime.png')
plt.show()
```

```python
#Set_index to Order Date//

df.set_index("Order Date", inplace=True)
```

```python
# Moving Averages (Trend Smoothing)
monthly_sales = df.groupby("order Month")["Sales"].sum().sort_index()
moving_avgs = monthly_sales.rolling(window=3).mean()

plt.figure(figsize=(14,6))
plt.plot(monthly_sales, label="monthly sales", alpha=0.5)
plt.plot(moving_avgs, label="3-month moving avg", linewidth=3)
plt.legend()
plt.title("Sales with moving averages")
plt.ylabel("Sales")
plt.xlabel("Months")
plt.grid(True)
plt.tight_layout()
plt.savefig('../images/Moving_AVGS.png')
plt.show()
```

## 💡 Key Insights

✅ Key Insights from EDA on E-commerce Superstore Dataset

📈 1. Sales and Profit Distribution

-Sales are right-skewed – most orders are low in value; few high-value orders exist.

-Profit shows extreme outliers — some transactions lead to high losses/gains.

-There’s a positive correlation (0.48) between Sales and Profit, but it's not very strong.

📦 2. Quantity and Discount

-Quantity is mostly between 2 and 6 items per order.

-Discount is negatively correlated with Profit (-0.22), implying higher discounts reduce profit.

🗺️ 3. Regional Performance

-West and East regions contribute the most to total orders.

-Central and South have fewer customers — potential for marketing focus.

📆 4. Time Series & Trend

-Seasonal spikes in November and December — possibly due to holidays/sales.

-Increasing trend overall when visualized with 3-month moving averages.

🧊 5. Correlation Matrix

-Moderate correlation between Sales & Profit (0.48).

-Negative correlation between Discount & Profit (-0.22) — deeper discount hurts profit.

-Almost no correlation between Quantity and Profit.



## 💼 Business Recommendations


Limit High Discounts:

-Since discounts negatively affect profit, implement a threshold-based discount strategy.

-Offer discounts only on low-performing or overstocked products.

Seasonal Promotions:

-Focus marketing and ad spend during Q4 (Oct–Dec) to maximize revenue when sales spike.

-Run campaigns in Q2 (Apr–Jun) to stabilize slower periods.

Expand in South and Central Regions:

-Explore local marketing or warehouse setup to increase presence in underperforming regions.

Product Bundling:

-Encourage sales of high-profit items with commonly bought low-profit items using bundles.

Monitor Outliers:

-Investigate transactions with extreme losses for operational inefficiencies or fraud.

## 📌 Conclusion


The E-commerce Superstore data analysis reveals a business with strong sales performance, but inconsistent profitability driven by high discounts and product-level variations. While sales are steadily increasing over time with strong peaks during the holiday seasons, profit margins do not always follow the same trend — especially in high-discount scenarios.

Regional and segment-based insights highlight opportunities for growth, particularly in the South and Central regions. Discounts, although helpful for driving sales, are shown to negatively impact profits, suggesting the need for more strategic discounting.
